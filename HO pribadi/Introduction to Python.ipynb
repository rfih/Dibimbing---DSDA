# **4. Latihan Santuy hehe**

1. coba buat input dengan variabel yang namanya suhu_C yang meminta user memasukan nilai suhu dalam celcius
2. ubahlah nilai suhu_C menjadi nilai suhu dalam fahrenheit
3. - Jika suhu Fahrenheit diatas 150, maka cetak â€œSekarang suhunya panasâ€
  - Jika suhu Fahrenheit diatas 100, maka cetak â€œSekarang suhunya normalâ€
  - Jika suhu Fahrenheit selain nilai diatas, maka cetak â€œSekarang suhunya dinginâ€

hint:
```
suhu_F = (0Â°C Ã— 9/5) + 32
```

````python
suhu_celcius = int(input("Masukkan suhu dalam celcius: "))
suhu_fahrenheit = (suhu_celcius * 9/5) + 32
if suhu_fahrenheit > 150:
  print(suhu_fahrenheit, "Sekarang suhunya panas")
elif suhu_fahrenheit > 100:
  print(suhu_fahrenheit, "Sekarang suhunya normal")
else:
  print(suhu_fahrenheit, "Sekarang suhunya dingin")
````
