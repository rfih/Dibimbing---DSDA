## 1. Distribusi data

adalah cara data diorganisasikan atau disebarkan dalam suatu dataset. Distribusi memberikan gambaran pola data, termasuk nilai tengah, penyebaran, dan kecondongan

contoh:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
number = [2,4,5,6,7,9,8]
mean = sum(number)/len(number)
mean
```

```python
import statistics
mean1 = statistics.mean(number)
mean1
```

```python
median = sorted(number)[len(number)//2]
median
```

```python
median1 = statistics.median(number)
median1
```

```python
mode = max(set(number), key=number.count)
mode
```

```python
mode1 = statistics.mode(number)
mode1
```

```python
std = statistics.stdev(number)
std
```

```python
count, bins, ignored = plt.hist(number, 30, density=True, alpha=.3)
plt.plot(bins, 1/(std * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mean1)**2 / (2 * std**2) ),
         linewidth=2, color='r')
lline = -.67*std + mean1
uline = .67*std + mean1
plt.axvline(lline, color='g')
plt.axvline(uline, color='g')
plt.title("Normal distribution with mean: {:.02f} and StDev: {:.02f}".format(mean1, std))
plt.show()
```

## 2. Perubahan Acak

---

Perubahan Acak adalah perubahan yang terjadi tanpa pola tertentu dan tidak dapat diprediksi secara pasti. Namun, dengan probabilitas kita dapat memodelkan kemungkinan hasilnya.

- Diskrit: Memiliki jumlah nilai yang terbatas atau dapat dihitung (contoh: jumlah kepala dalam 3 lemparan koin).
- Kontinu: Memiliki nilai yang terhingga dalam rentang tertentu (contoh: tinggi badan orang).

Ciri - ciri peubah acak
- Nilai peubah acak terkait dengan probabilitas tertentu.
Dapat direpresentasikan dengan PMF untuk peubah diskrit atau PDF unutk peubah kontinu.

PDF : probability density function
PMF : Probability mass function

## 3. Probability Mass Function (PMF)

---
PMF adalah fungsi yang menggambarkan probabilitas variabel diskrit berada pada nilai tertentu.

Sifat PMF
- Probabilitas selalu bernilai 0 dan 1
- Total nilai akan menghasilkan nilai 1
( 1/6+1/6+1/6+1/6+1/6+1/6 = 6/6 atau 1 )



## 4. Probability Density Function (PDF)

---

PDF adalah fungsi yang menggambarkan probabilitas untuk variable kontinu dan area di bawah kurva PDF merepresentasikan probabilitas.

Sifat PDF
- Selalu bernilai non-negatif
- Area total dibawah kurva adalah 1

Contoh PDF
- Distribusi normal: Tinggi badan orang dewasa dapat dimodelkan dengan PDF distribusi normal. Probabilitas tinggi bana seseorang dalam rentang tertentu, misalnya antara 160 cm hingga 180 cm

## 7. Z-Score

---

Z-Score adalah nilai yang menunjukkan seberapa jauh data berada dari mean dalam satuan standar deviasi

Rumus:

z = (x − μ) / σ ‍

x = observed value
μ = mean
σ = stdeviation

interpretasi z-score:
- z=0 : Data berada tepat di mean
- z>0 : Data berada di atas mean.
- z<0 : Data berada di bawah mean.

Aplikasi Z-Score
- Mengidentifikasi outlier.
- Membandingkan nilai antar dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import stats
```

```python
data = np.random.rand(1000)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, bins=30)
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data, fill=True) #density plot (kdeplot)
plt.title('Density Plot of Sample Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Normal Distribution
data1 = np.random.normal(loc=0, scale=1, size=1000) # Mean=0, SD=1
data2 = np.random.normal(loc=3, scale=0.5, size=1000) # Mean=3, SD=0.5
data3 = np.random.normal(loc=-2, scale=2, size=1000) # Mean=-2, SD=2

# Create a DataFrame for easier plotting
df = pd.DataFrame({
'Standard Normal': data1,
'Normal(μ=3, σ=0.5)': data2,
'Normal(μ=-2, σ=2)': data3
})

# Plot histograms with density curves
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns):
 sns.histplot(df[column], kde=True, stat='density', alpha=0.4, label=column)
plt.title('Comparison of Normal Distributions with Different Parameters')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
