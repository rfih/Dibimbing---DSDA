### Basic Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris, load_wine, load_diabetes

# Set styling for plots
plt.style.use('seaborn-v0_8-whitegrid')
sns.set(style="whitegrid")

# Load data
data = load_iris()
data_df = pd.DataFrame(data.data, columns=data.feature_names)
data_df['species'] = pd.Categorical.from_codes(data.target, data.target_names)
data_df
```

```python
### Scatter Plot
plt.figure(figsize=(10,6))
plt.scatter(data_df['sepal length (cm)'], data_df['sepal width (cm)'])
plt.title('Sepal Length vs Sepal Width (CM)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
### Seaborn Enhanced Scatter Plot
plt.figure(figsize=(12,8))
sns.scatterplot(
    data = data_df,
    x = 'sepal length (cm)',
    y = 'sepal width (cm)',
    hue = 'species',
    palette = 'viridis'
)

plt.title('Sepal Dimensions by Species', fontsize=15)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize = 12)
plt.legend(title='species', fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(
    data=data_df,
    x='petal length (cm)',
    y='petal width (cm)',
    hue='species',
    palette = 'cubehelix',
    alpha = 1
)

plt.title('Petal Dimension by Species', fontsize =15)
plt.xlabel('Petal length', fontsize = 15)
plt.ylabel('Petal width', fontsize = 15)
plt.legend(title = 'species', fontsize = 15)
plt.show()
```

### Spotify Visualization

```python
data = pd.read_csv('spotify_dataset.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
### Popularity Distribution
plt.figure(figsize=(12,10))
sns.histplot(data = data, x='popularity', kde=True)
plt.title('Popularity by Distribution')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

```python
data['popularity'].describe()
```

### Insight :
1. mean < median < modus (left skew)
2. Distribusi popularity berada di rentang angka 40-80
3. Terdapat nilai popularity 0 sekitar 175
4. Modus pada popularity berada di nilai 70

```python
## Line Chart (Hitung total lagu berdasarkan tahun)
plt.figure(figsize=(10,8))
sns.lineplot(data=data, x='year', y='song')
plt.title('Total Song by Year')
plt.xlabel('Year')
plt.ylabel('Total Song')
plt.show()
```

```python
data_year = data[['year', 'song']]
data_year
```

```python
data_yearly = data_year.groupby('year').count()
data_yearly
```

```python
## Line chart total song per year
plt.figure(figsize=(10,8))
sns.lineplot(data=data_yearly, x='year', y='song')
plt.title('Total Song by Year')
plt.xlabel('Year')
plt.ylabel('Total Song')
```

### Insight :       
1. 2001 hingga 2018 jumlah lagu berada di posisi yang stabil yang berada di angka 85-115
2. Ditahun 2012 merupakan tahun tertinggi pada saat release lagu di spotify

```python
### Bar chart (Perhitungan lagu tiap genre)
data_genre = data.groupby('genre', as_index=False).count().sort_values(by='song', ascending=False)
data_genre_top10 = data_genre.head(10)

plt.figure(figsize=(10,8))
sns.barplot(data=data_genre_top10, x='genre', y='song')
plt.title('Top 10 Song by Genre')
plt.xlabel('Genre')
plt.xticks(rotation=90)
plt.ylabel('Total Song')
plt.show()
```

### Insight :    
1. Lagu pop yang paling banyak diminati, disusul dengan lagu hip hop
2. Genre Dance/electronic kurang diminati

```python
### Pie Chart (Explicit distribution)
explicit_data = data.groupby('explicit', as_index=False).count()

plt.figure(figsize=(10,8))
plt.pie(explicit_data['song'], labels=explicit_data['explicit'], autopct='%1.1f%%')
plt.title('Explicit Distribution')
plt.show()
```

### Insight :       
1. Lirik lagu atau video musik yang mengandung satu atau lebih kriteria yang dapat menyinggung  atau tidak cocok untuk anak-anak lebih sedikit dengan porsi 27,5%

```python
### Scatter Plot (Duration and Year)
plt.figure(figsize=(12,10))
sns.scatterplot(data=data, x='year', y='duration_ms')
plt.title('Duration vs Year')
plt.xlabel('Year')
plt.ylabel('Duration (ms)')
plt.show()
```

```python
### Correlation between duration and year
data[['duration_ms', 'year']].corr()
```

### Insight : Semakin bertambah tahun, maka durasi lagu semakin berkurang (weak negatif correlation)

## Day 14 - Advanced Data Visualization

```python
data
```

```python
### Correlation
numeric_columns = data.select_dtypes(include=['number']).columns
numeric_columns
```

```python
corr_matrix = data[numeric_columns].corr()
corr_matrix
```

```python
### Heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='viridis')
plt.title('Correlation Matrix')
plt.show()
```

### Insight :      
1. Variabel year dan duration (ms) berkorelasi negatif cenderung kuat (-0.32), yang artinya semakin tinggi atau besar tahunnya semakin pendek durasinya (ms) dan sebaliknya
2. Variabel accousticness memiliki korelasi negatif kuat (-0.45) dengan variabel energy, yang artinya semakin accoustic lagunya maka semakin rendah energynya dan sebaliknya
3. Variable danceability memiliki korelasi positif kuat (0.40) dengan variabel valence, yang berarti semakin tinggi danceability maka semakin besar valence
4. Variabel loudness dan variabel energy memiliki korelasi positif yang sangat kuat (0.65) yang artinya semakin tinggi loudness maka semakin besar energy

```python
## Line chart with fills
data_yearly = data_year.groupby('year').count()
data_yearly

plt.figure(figsize=(10,8))
plt.style.use('dark_background')
sns.lineplot(data=data_yearly, x='year', y='song', color='blue', marker = 'o')

# setting background
plt.gca().set_facecolor('darkgrey')

plt.title('Total Songs by Years')
plt.xlabel('Year')
plt.ylabel('Song', fontweight='bold')
plt.grid(True)
```

```python
### Menambahkan fills dibawah line chart
plt.figure(figsize=(10,8))
plt.style.use('dark_background')
sns.lineplot(data=data_yearly, x='year', y='song', color='blue', marker='o', linestyle ="--")

# setting background
plt.gca().set_facecolor('darkgrey')
plt.fill_between(data_yearly.index, data_yearly['song'], color='red', alpha=0.2)

plt.title('Total Songs by Years')
plt.xlabel('Year')
plt.ylabel('Song', fontweight='bold')
plt.grid(True)
```

```python
### Menambahkan Label pada setiap titik
plt.figure(figsize=(10,8))
plt.style.use('dark_background')
sns.lineplot(data=data_yearly, x='year', y='song', color='blue', marker='o')

# setting background
plt.gca().set_facecolor('darkgrey')
plt.fill_between(data_yearly.index, data_yearly['song'], color='red', alpha=0.2)

# menambahkan angka pada setiap titik
for x,y in zip(data_yearly.index, data_yearly['song']):
    plt.text(x, y, str(y), ha='center', va='bottom', color='yellow')


plt.title('Total Songs by Years')
plt.xlabel('Year')
plt.ylabel('Song', fontweight='bold')
plt.grid(True)
```

```python
### Advanced Bar Chart
data_genre = data.groupby('genre', as_index=False).count().sort_values(by='song', ascending=False)
data_genre_top10 = data_genre.head(10)

plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(10,8))
barplot = sns.barplot(data=data_genre_top10, x='genre', y='song', palette='crest')

## Tambahkan label di setiap bar chart
for index, value in enumerate(data_genre_top10['song']):
  plt.text(index, value, str(value), ha='center', va='bottom', fontsize=11, color='black')
plt.title('Top 10 Song by Genre')
plt.xlabel('Genre')
plt.xticks(rotation=90)


sns.despine()
plt.tight_layout()
plt.show()
```

```python
# Data Churn
data_churn = pd.read_csv('Telco_customer_churn.csv')
data_churn
```

```python
## Churn reasons (bar chart)
churn_reason = data_churn['Churn Reason'].value_counts().sort_values(ascending=False)
churn_reason
```

```python
churn_reason = churn_reason.reset_index()
churn_reason
```

```python
plt.figure(figsize=(10,8))
plt.style.use('default')
sns.barplot(data=churn_reason, x='Churn Reason', y='count', palette='crest')
sns.set_context('notebook')

## Set titles and labels
plt.title('Churn Reasons', fontweight='bold')
plt.suptitle('Churn Analysis', fontweight='bold')
plt.xlabel('Reason')
plt.ylabel('Count')
plt.xticks(rotation=90, horizontalalignment='right')
plt.gca().set_facecolor('grey')
```

```python
### Multiple bar chart
internet_service = data_churn.groupby(['Internet Service', 'Churn Label']).size().reset_index(name='Jumlah')
internet_service
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(data = internet_service, x='Internet Service', y='Jumlah', hue='Churn Label', palette = {'No':'Red', 'Yes':'Green'})

# Menambahkan label di setiap bar
for s in ax.containers:
  ax.bar_label(s, fmt='%.0f')

plt.title('Comparison Internet Service and Churn')
plt.xlabel('Internet Service')
plt.ylabel('Jumlah')
plt.legend(title='Churn Label')
plt.xticks(rotation=90)
plt.show()
```
