## Playing around with normal distribution

```python
# IQ follow Normal (100,115)
# Probability IQ between 100 and 115?
dist = stats.norm(loc=100, scale=15)
prob = dist.cdf(115) - dist.cdf(100)
print(f"Probability IQ between 100 and 115 is {prob:.4f}")
```

```python
#get the same probability from Z score
z_score_kanan = (115 - 100) / 15 #1
z_score_kiri = (100 - 100) / 15 #0

# pr (mean < X < mean+std)
prob = stats.norm.cdf(z_score_kanan) - stats.norm.cdf(z_score_kiri)
print(f"Probability IQ between 100 and 115 is {prob:.4f}")
```

# T-Test

## Transaction Amount Data

```python
# snippet of data
trx_amount_df.head()
```

```python
# general info
trx_amount_df.info()
```

```python
trx_amount_df.describe()
```

From the mean values: variant > control. Is this difference significant?

```python
# T-test
from statsmodels.stats.weightstats import ttest_ind
df = trx_amount_df
t_stat, p_val, _ = ttest_ind(df["variant"], df["control"])

print(f" T-statistic: {t_stat:.4f}")
print(f" P-value: {p_val:.4f}")
```

## Payment Duration Data

```python
payment_duration_df.head()
```

```python
payment_duration_df.describe()
```

From the mean values: variant > control. Is this difference significant?

```python
# t-test
from statsmodels.stats.weightstats import ttest_ind
df = payment_duration_df
t_stat, p_val, _ = ttest_ind(df["variant"], df["control"])

print(f" T-statistic: {t_stat:.4f}")
print(f" P-value:{p_val:4f}")
```

p-value < 0.05 -> reject H0: result is significant

```python
df_trx_abc.head()
```

```python
# preprocess data
df_trx_abc_melt = df_trx_abc.melt(id_vars=None,
                                  var_name='experiment_group',
                                  value_name='trx_amount')
df_trx_abc_melt.head()
```

```python
df_trx_abc_melt.value_counts('experiment_group')
```

```python
# one-way ANOVA using pingouin library
anova_res = df_trx_abc_melt.anova(dv="trx_amount", between= "experiment_group", detailed=True)
anova_res.loc[:,["Source","SS","DF","MS","F","p-unc"]]
```

COnclusion: Reject H0! There is at least one group that has differenct transaction amount average!
