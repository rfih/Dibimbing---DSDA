## 1. Introduction to Data Visualization

---

Visualisasi data adalah representasi grafis dari informasi dan data dengan menggunakan elemen visual seperti bagan, grafik dan peta.

Tujuan visualisasi data adalah memudahkan pembaca untuk melihat dan memahami tren, outlier dan pola dalam data.


## 2. Visualizing Bivariate Data

---
1. Line Chart

Tujuan Penggunaan dari visualisasi line chart
- Melacak perubahan dari waktu ke waktu menggunakan interbal waktu yang sama
- mendeteksi pola musiman atau siklus dalam data

2. Scatter Plot

Tujuan penggunaan dari visualisasi scatter plot
- mengidentifikasi apakah ada hubungan atau korelasi antara dua variabel numerik.
- menguji hipotesis tentang hubungan antara dua variabel.



#Hands Up

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris, load_wine, load_diabetes

# Set styling for plots
plt.style.use('seaborn-v0_8-whitegrid')
sns.set(style="whitegrid")

# Load data
data = load_iris()
data_df = pd.DataFrame(data.data, columns=data.feature_names)
data_df['species'] = pd.Categorical.from_codes(data.target, data.target_names)
data_df
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data_df['sepal length (cm)'], data_df['sepal width (cm)'])
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# Seaborn enhanced scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', palette='viridis', alpha = 0.7, data=data_df)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.legend(title='Species')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_df['petal length (cm)'], data_df['petal width (cm)'])
plt.title('Scatter Plot of Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='species', palette='viridis', alpha = 0.7, data=data_df)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Scatter Plot of Petal Length vs Petal Width')
plt.legend(title='Species')
plt.show()
```

## Spotify Visualization

```python
data = pd.read_csv('spotify_dataset.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
# Popularity Distribution
plt.figure(figsize=(10,6))
sns.histplot(data = data, x='popularity', kde=True)
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

Left Skewed (negative Skewed)

```python
data['popularity'].describe()
```

Insight :    
1. mean < median < modus (left skew)
2. Distribusi popularity berada di rentang angka 40-90
3. Terdapat nilai popularity 0 sekitar 175
4. Modus pada popularity berada di nilai 70

## Aggregation

```python
# Line chart (hitung total lagu berdasarkan tahun)
plt.figure(figsize=(10,6))
sns.lineplot(data=data, x='year', y='song')
plt.title('Total Songs by Year')
plt.xlabel('Year')
plt.ylabel('Total Songs')
plt.show()
```

```python
data_year = data[['year', 'song']]
data_year
```

```python
data_yearly = data_year.groupby('year').count()
data_yearly
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=data_yearly, x='year', y='song')
plt.title('Total Songs by Year')
plt.xlabel('Year')
plt.ylabel('Total Songs')
plt.show()
```

Insight:
1. 2001 hingga 2018 jumlah lagu berada diposisi stabil berada di sekitar angka 85 - 115
2. Ditahun 2012 merupakan tahun tertinggi pada saat release lagu di spotify

## Barchart

```python
# Barchart (Perhitungan lagu tiap genre)
data_genre = data.groupby('genre', as_index=False).count().sort_values(by='song', ascending=False)
data_genre_top10 = data_genre.head(10)
data_genre_top10
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=data_genre_top10, x='genre', y='song')
plt.title('Total Songs by Genre')
plt.xlabel('Genre')
plt.xticks(rotation=90)
plt.ylabel('Total Songs')
```

Insight:
1. Lagu pop paling banyak diminati disusul dengan lagu hip hop.
2. genre dance/electronic kurang diminati.

## Piechart

```python
# Piechart (Explicit distribution)
explicit_data = data.groupby('explicit', as_index=False).count()

plt.figure(figsize=(10,6))
plt.pie(explicit_data['song'], labels=explicit_data['explicit'], autopct='%1.1f%%')
plt.title('Explicit Distribution')
plt.show()
```

Insight:
1. Lirik lagu atau video music yang mengandung satu atau lebih kriteria yang dapat menyinggung atau tidak cocok untuk anak-anak lebih sedikit dengan porsi 27.5%

```python
### Scatter Plot (Duration and Year)
plt.figure(figsize=(12,10))
sns.scatterplot(data=data, x='duration_ms', y='year')
plt.title('Duration vs Year')
plt.xlabel('Duration (ms)')
plt.ylabel('Year')
plt.show()
```

```python
# correlation between duration and year
data[['duration_ms', 'year']].corr()
```

Insight:

Semakin bertambah tahun, maka durasi semakin berkurang (weak negatif correlation)
