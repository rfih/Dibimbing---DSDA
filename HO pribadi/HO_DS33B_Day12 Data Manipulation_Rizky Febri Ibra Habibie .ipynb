# Join/Merge

## Left Table: Tips

```python
# There are many methods that we can use for grouping. The method '.size()'
# is like the count(*) in SQL. So, the following code is similar to  this SQL

# SELECT size, COUNT(*) FROM tips GROUP BY size;

df_tips.groupby('size').size()
#df_tips.groupby('time').agg({'time': 'count'})
```

```python
# We're going to save the result into a new dataframe to be used later for joins

df_tips_size = df_tips.groupby('size').size().reset_index()
df_tips_size.columns = ['size', 'total_order']
df_tips_size.head()
```

## Right Table Size

```python
sizes = { 2: 'Short', 3: 'Medium', 4: 'Tall', 5: 'Grande', 6: 'Venti' , 7: 'Quaranta'}

df_sizes = pd.DataFrame(sizes.items(), columns=['size', 'size_name'])
df_sizes
```

## Inner Join

```python
pd.merge(
    df_tips_size,
    df_sizes,
    how='inner',
    left_on='size',
    right_on='size'
)
```

```python
# If the key column from the left table has the same name with the key column
# from the right table, we can use parameter 'on' instead.
pd.merge(
    df_tips_size,
    df_sizes,
    how='inner',
    on='size',
)
```

## Left Join

```python
pd.merge(
    df_tips_size,
    df_sizes,
    how='left',
    on='size',
)
```

## Right Join

```python
pd.merge(
    df_tips_size,
    df_sizes,
    how='right',
    on='size',
)
```

## Full Outer Join

```python
pd.merge(df_tips_size, df_sizes, how='outer', on='size')
```

## Discretization

```python
  df_weather['temperature'].plot.hist()
```

```python
import numpy as np
```

```python
# One way to implement if-else in pandas.
#
# if temperature < 0:
#     '[1] Below Zero'
# elif temperature < 16:
#     '[2] Between 0 and 16'
# elif temperature < 27:
#     '[3] Between 16 and 27'
# else:
#     '[4] Over 27'
#
np.where(
    df_weather['temperature'] < 0,
    '[1] Below Zero',
    np.where(
        df_weather['temperature'] < 16,
        '[2] Between 0 and 16',
        np.where(
            df_weather['temperature'] < 27,
            '[3] Between 16 and 27',
            '[4] Over 27'
        )
    )
)
```

```python
# Or, we can use binning.

# Discretize into four equal-sized bins.
# NOTE: In practice, we don't need to use the parameter 'retbins'. We use it
# here to show the intervals.
pd.cut(df_weather['temperature'], 4, retbins=True)
```

```python
# Discretize into four equal-sized bins with labels.
labels = ['coldest', 'cold', 'warm', 'warmer']
pd.cut(df_weather['temperature'], 4, labels=labels, retbins=True)
```

```python
# Discretize into custom bins.

bins = [-5, 0, 16, 27, 32]

pd.cut(df_weather['temperature'], bins=bins, retbins=True)
```

```python
# Discretize into custom bins.
labels = [
    '[1] Below Zero',
    '[2] Between 0 and 16',
    '[3] Between 16 and 27',
    '[4] Over 27',
]

bins = [-5, 0, 16, 27, 32]

pd.cut(df_weather['temperature'], bins=bins, labels=labels, retbins=True)
```

## Min - Max

```python
# The vanilla way.

pressure_min = df_weather['pressure'].min()
pressure_max = df_weather['pressure'].max()

df_weather['pressure_normalized_min_max'] = \
    (df_weather['pressure'] - pressure_min) / (pressure_max - pressure_min)
```

```python
df_weather['pressure_normalized_min_max'].plot.hist()
```

```python
# The sklearn way.
# Why would I want to use sklearn if there's an easier way?
```

##Z-Score (Standardization)

```python
# The vanilla way.

pressure_mean = df_weather['pressure'].mean()
pressure_std = df_weather['pressure'].std()

df_weather['pressure_normalized_z_score'] = \
    (df_weather['pressure'] - pressure_mean) / pressure_std
```

```python
df_weather['pressure_normalized_z_score'].plot.hist()
```
