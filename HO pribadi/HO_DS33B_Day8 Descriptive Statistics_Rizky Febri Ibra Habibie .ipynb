# 5 Normal Distribution

```python
from scipy import stats
```

## 5.1 Normal Distribution

```python
# IQ follow Normal(100,15)
# Probability IQ between 120 and 140?
dist = stats.norm(loc=100, scale=15)
```

```python
dist.cdf(140)
```

```python
dist.cdf(120)
```

```python
prob = dist.cdf(140) - dist.cdf(120)
print(f"Probability IQ between 120 and 140: {prob*100:.2f} %")
```

## 5.2 Standard Normal Distribution

```python
# get z-score from alpha
# alpha to critical
alpha = 0.05
n_sided = 2 # in practice, it's (almost) always 2 sided test
z_crit = stats.norm.ppf(1-alpha/n_sided)
print(f"Z-critical value for alpha = 0.05 is {z_crit:.2f}")
```

# 6 Expected Value

```python
bonus = [0,50_000, 100_000] # Nilai yang mungkin didapat
probabilities = [0.6, 0.3, 0.1] # Probabilities masing-masing nilai
```

```python
#Hitung expected value
expected_value = sum(x * p for x, p in zip(bonus, probabilities))
print(f"Expected Value bonus bulanan: {expected_value}")
```

```python
print(f"Expected Value bonus bulanan: Rp {expected_value*12:,.0f}")
```

```python
#Misal data pembelian tiket per bulan selama 12 bulan untuk satu agen:
ticket_purchases = [0,1,2,3,4,5,6,7,8]
frequencies = [1,1,3,2,1,1,1,1] #frekuensi (jumlah bulan)
```

```python
total_months = sum(frequencies)
probabilities_2 = [f / total_months for f in frequencies]
```

```python
#Hitung expected value untuk pembelian tiket agen
expected_value_tickets = sum(x * p for x, p in zip(ticket_purchases, probabilities_2))
print(f"Expected purchases per bulan: {expected_value_tickets:.2f}")
```
